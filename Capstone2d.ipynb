```python
toronto_data= dfa[dfa['Borough'].str.contains('Toronto', na = False)].reset_index(drop=True)
toronto_data.head()
```

```python
toronto_data1 = toronto_data['Neighborhood'].str.replace("/",",")
```

```python
toronto_data['Neighborhood'] = toronto_data1
```

```python
toronto_data2 = toronto_data[toronto_data['Borough'].str.contains('Toronto', na = False)].reset_index(drop=True)
toronto_data2.head()
```

From google : the latitud and longitud for Toronto, as follows: 43.6532° N, 79.3832° W

```python
latitude = 43.6532
longitude= -79.3832
```

```python
!conda install -c conda-forge folium --yes
import folium # map rendering library
```

```python
# create map of TORONTO using latitude and longitude values above:
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data2['Latitude'], toronto_data2['Longitude'], toronto_data2['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
